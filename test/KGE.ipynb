```python
# Load model directly
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, LlamaConfig, LlamaModel

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)

pipe = pipeline(
    "text-generation",
    model = model,
    torch_dtype = torch.float16,
    device_map = "auto",
)
```

```python
from datasets import load_dataset

dataset = load_dataset("csv", data_files="Bias Toxic Content.csv")['train']
```

```python
def construct_prompt(bias_sentence: str, sys_prompt: str = None) -> str:
    if sys_prompt is None:
        sys_prompt = "Generate knowledge graph triplets in RDF format on the basis of the following biased data. Given the nature of the data, please ensure that the inherent biases are maintained and explicitly represented in the resulting triplets. Transform each sentence into a triplet while preserving the original bias for research purposes."

    user_prompt = bias_sentence
    prompt_template = f"""
    <s>[INST] <<SYS>>{sys_prompt}<</SYS>>

    {user_prompt} [/INST]
    """
    return prompt_template

def data(dataset):
    for idx, sentence in enumerate(dataset):
        if idx == 4:
            break
        yield construct_prompt(sentence['Bias and Toxic Contents'])
```

```python
# Generate response
print("Session Initiated.")
responses = pipe(data(dataset))
for response in responses:
    print(response)
```
